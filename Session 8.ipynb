```
def averager():
    total = 0
    count = 0
    def add(number):
        nonlocal total
        nonlocal count
        total = total + number
        count = count + 1
        return total/count
    return add
```

```
def counter(fn):
    cnt = 0
    def inner(*args, **kwargs):
        nonlocal cnt
        cnt += 1
        print('{0} has been called {1} times'.format(fn.__name__, cnt))
        return fn(*args, **kwargs)
    return inner
```

```
def add(a, b):
    return a + b
def mul(a, b):
    return a*b
```

```
counter_add = counter(add)
```

```
counter_add.__closure__
```

```
counter_add(1, 2)
```

```
counter_add(10, 20)
```

```
counter_mult = counter(mul)
counter_mult(2, 5)
counter_mult(2, 5)
counter_mult(2, 5)
counter_mult(2, 5)
counter_mult(2, 5)
```
